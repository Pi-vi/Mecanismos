Sistemas came seguidor são frequentemente usados em todos os tipos de máquinas. As válvulas do motor do seu automóvel são operadas por cames. Máquinas usadas na manufatura de bens de consumo estão cheias de cames. 
* Comparado aos mecanismos de barras, cames são mais simples para projetar uma função específica como saída, porém são mais difíceis e mais caros de construir do que um mecanismo de barras. 
* Para qualquer posição instantânea do came e do seguidor, podemos substituir por um mecanismo de barras que, para essa posição instantânea, executará o mesmo movimento que o original. 
* Podemos virtualmente especificar qualquer função de saída que desejarmos e criar uma superfície curva no came para gerar aquela função no movimento do seguidor. 

## 8.1 Terminologia para Cames

Sistemas came seguidor podem ser classificados de muitas maneiras:
* Pelo tipo de movimentação do seguidor:
    * de translação
    * de rotação (oscilação)
* pelo tipo de came:
    * radial
    * rolete
    * tridimensional
* pelo tipo de fechamento da junta:
    * de força
    * de forma
* pelo tipo de seguidor:
    * curvo ou liso
    * rotacionando ou deslizando
* pelo tipo de movimento de crítico
    * posição extrema crítica (PEC)
    * percurso de movimento crítico (PMC)
* pelo tipo programado de movimentação
    * sobe-desce (SD)
    * sobe-desce-para (SDP)
    * sobe-para-desce-para (SPDP)

### Tipo de movimentação do seguidor

* de translação
* de rotação (oscilação)

A escolha entre essas duas formas de came seguidor é usualmente ditada pelo tipo de movimentação desejada na saída.

### Tipo de fechamento da junta

* A **junta de força** requer que uma força externa seja aplicada na junta para manter os dois elos, came e seguidor, fisicamente em contato. 
* A **junte de forma** força o contato nos elos por geometria.

### Tipos de restrições de movimento

* posição extrema crítica (PEC, também conhecida como especificação de ponto final)
* percurso de movimento crítico (PMC)

**Posição extrema crítica** refere-se ao caso em que especificações do projeto definem a posição inicial e final do seguidor (isto é, posições extremas), mas não especificam qualquer restrição no percurso entre as duas posições extremas.

**Percurso de movimento crítico** é um caso mais complicado que o PEC, porque o percurso de movimento e/ ou uma ou mais de suas derivadas precisam ser definidos em todo ou em parte do intervalo do movimento. Isso é análogo à **geração de função** no projeto de mecanismo de barras, exceto que, com o came, podemos obter uma função de saída contínua para o seguidor.

### Tipo de programa de movimentação

Os progrmas de movimentação **sobe-desce** (SD), **sobe-desce-para** (SDP) e **sobe-para-desce-para** (SPDP) referem-se todos principalmente ao caso PEC de restrição de movimentação e, na verdade, definem quantas esperas são apresentadas no ciclo completo de movimentação: nenhuma (SD), um (SDP) ou mais de uma (SPDP). 

## 8.2 Diagramas E V A P

A primeira tarefa do projetista de came é selecionar a função matemática a ser usada para definir o movimento do seguidor. 
* O caminho mais fácil é "linearizar" o came, isto é, desenvolvê-lo a partir de sua forma circular e considerá-lo uma função desenhada nos eixos cartesianos.

    * sua segunda derivada a aceleração $a$; 
    * e sua terceira derivada, o pulso $p$.

## 8.3 Projeto do Came com Dupla Espera - Escolhendo as Funções E V A P

Muitas aplicações de projeto do came requerem esperas múltiplas. O caso da espera dupla é muito comum. Suponha que um came com **dupla espera** esteja movendo uma parte da estação de alimentação em uma máquina de produção de pasta de dentes. Esse hipotético came seguidor é carregado com um tubo de pasta de dentes vazio (durante uma espera inferior), então move o tubo vazio para a estação de carregamento (durante a subida), mantém o tubo absolutamente fixo em uma **posição extrema crítica** (PEC), enquanto a pasta de dentes é espremida para dentro da parte aberto do tubo (durante a espera superior), e então o tubo preenchido é levado de volta à posição inicial (zero) e mantido na outra posição extrema crítica. Nesse ponto, outro mecanismo (durante a espera inferior) agarra o tubo e o leva até a próxima operação, que deve selar o fundo do tubo. Um came similar pode ser usado também para alimentar, alinhar e retirar o tubo da estação onde se sela o fundo.

* O **ciclo da máquina** é definido como uma revolução de seu eixo motor. 
* As especificações do funcionamento do came são posicionadas em uma **diagrama de tempo**.
* Existe um diagrama de tempo para cada submontagem da máquina.
* As relações de tempo entre todas as submontagens são definidas por seus diagramas de tempo.

![](fig8-7.jpg)

### A lei fundamental de projeto do came

* A função do came deve ser contínua por toda primeira e segunda derivada do deslocamento dutante todo o intervalo (360 graus).

#### Colorário

* A função pulso deve ser finita durante todo o intervalo (360 graus).

## Projeto do Came com Dupla Espera - Escolhendo as Funções EVAP

Para este projeto utilizaremos os seguintes dados: 
* **espera** - no deslocamento zero por 90 graus
* **subida** - 25mm em 90 graus
* **espera** - em 25mm por 90 graus (espera superior)
* **descida** - 25mm em 90 graus 
* **came** $\boldsymbol{\omega}$ - $2\pi$ rad/s - 1rev/s

**Primeira Tentativa**

```python
from sympy import *
```

```python
theta,Kv = symbols('theta K_v')
```

```python
s = Kv*theta
s
```

Onde Kv é a velocidade constante

```python
v = s.diff(theta)
v
```

```python
a = v.diff(theta)
a
```

```python
j = a.diff(theta)
j
```

**Segunda tentativa - movimento harmônico**

```python
h,beta= symbols('h beta')
```

```python
s = (h/2)*(1- cos(pi*theta/beta))
s
```

onde $h$ é a subida total, ou elevação; $\theta$ é o eixo do ângulo do came; e $\beta$ é o ângulo tota do intervalo de subida.

```python
v = s.diff(theta)
v
```

```python
a = v.diff(theta)
a
```

```python
j = a.diff(theta)
j
```

* O tempo de qualqer segmento é definido pelo ângulo $\beta$ e o seu valor pode ser diferente em cada segmento.
* o valor de $\theta/\beta$ irá variar de 0 a 1 em qualquer segmento

```python
dic1 = {h:25,beta:90}
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta_vals = np.linspace(0,90,90)
```

```python
lam_s = lambdify(theta,s.subs(dic1),modules = ['numpy'])
s_vals = lam_s(theta_vals)
plt.figure(dpi=150)
plt.plot(theta_vals,s_vals)
plt.xlabel(r'$\theta$[graus]')
plt.ylabel(r's($\theta$)')
plt.grid(True)
```

```python
lam_v = lambdify(theta,v.subs(dic1),modules = ['numpy'])
v_vals = lam_v(theta_vals)
plt.figure(dpi=150)
plt.plot(theta_vals,v_vals)
plt.xlabel(r'$\theta$[graus]')
plt.ylabel(r'v($\theta$)')
plt.grid(True)
```

```python
lam_a = lambdify(theta,a.subs(dic1),modules = ['numpy'])
a_vals = lam_a(theta_vals)
plt.figure(dpi=150)
plt.plot(theta_vals,a_vals)
plt.xlabel(r'$\theta$[graus]')
plt.ylabel(r'a($\theta$)')
plt.grid(True)
```

```python
lam_j = lambdify(theta,j.subs(dic1),modules = ['numpy'])
j_vals = lam_j(theta_vals)
plt.figure(dpi=150)
plt.plot(theta_vals,j_vals)
plt.xlabel(r'$\theta$[graus]')
plt.ylabel(r'j($\theta$)')
plt.grid(True)
```

```python
espera1 = np.zeros(90)
espera2 = 25*np.ones(90)
th1 = np.linspace(0,90,90)
th2 = np.linspace(90,180,90)
th3 = np.linspace(180,270,90)
th4 = np.linspace(270,360,90)
```

```python
plt.figure(dpi=150)
plt.plot(th1,espera1)
plt.plot(th2,s_vals)
plt.plot(th3,espera2)
plt.plot(th4,25-s_vals)
plt.xlabel(r'$\theta$ graus')
plt.ylabel('s')
plt.grid(True)
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(th1,espera1)
plt.plot(th2,v_vals)
plt.plot(th3,espera1)
plt.plot(th4,-v_vals)
plt.xlabel(r'$\theta$ graus')
plt.ylabel('v')
plt.grid(True)
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(th1,espera1)
plt.plot(th2,a_vals)
plt.plot(th3,espera1)
plt.plot(th4,-a_vals)
plt.xlabel(r'$\theta$ graus')
plt.ylabel('a')
plt.grid(True)
plt.show()
```

A função de aceleração não é contínua, as funções de espera, que unem a subida em cada lado, têm acelerações nulas.

**Terceira tentativa - Deslocamento cicloidal**

Os dois exemplos ruins de projeto de came descritos acima deveriam levar o projetista de came a concluir que considerar somente a função do deslocamento quando se projeta um came é errado. A melhor abordagem é começar considerando as funções da aceleração e do pulso. 

Dessa vez, começaremos pela função da aceleração. Observa-se que a função seno, em um ciclo de $0$ a $2\pi$, tem o seu início e término em zero. 

```python
C = symbols('C')
```

C é um símbolo que definirá a amplitude da curva seno

```python
a = C*sin(2*pi*theta/beta)
a
```

```python
v = integrate(a,theta)
v
```

```python
print(v)
```

```python
k1 = symbols('k1')
v = C*((-beta*cos(2*pi*theta/beta)/(2*pi))) + k1
v
```

Temos a condição de contorno $v=0$ em $\theta=0$, pois devemos igualar a velocidade nula do tempo de espera naquele ponto.

```python
v.subs(theta,0)
```

```python
k1_sol = solve(v.subs(theta,0),k1)[0]
k1_sol
```

```python
v = v.subs(k1,k1_sol)
v
```

```python
s = integrate(v,theta)
s
```

```python
print(s)
```

```python
k2 = symbols('k2')
s = C*beta*theta/(2*pi) - C*beta*((beta*sin(2*pi*theta/beta)/(2*pi)))*1/(2*pi) + k2
s
```

Temos a condição de contorno $s=0$ em $\theta=0$:

```python
k2_sol = solve(s.subs(theta,0),k2)[0]
k2_sol
```

```python
s = s.subs(k2,k2_sol)
s
```

Para validar a constante $C$ da amplitude, basta substituir $s=h$ em $\theta=\beta$:

```python
C_sol = solve(s.subs(theta,beta) - h,C)[0]
C_sol
```

Substituindo o valor da constante $C$ em $a$:

```python
a = a.subs(C,C_sol)
a
```

```python
v = v.subs(C,C_sol)
v
```

```python
s = s.subs(C,C_sol)
s
```

```python
j = a.diff(theta)
j
```
